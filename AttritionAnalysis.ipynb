## Data Transformation

In this analysis, I’ll focus on a subset of the dataset, since our primary goal is to examine employee attrition within the company. I anticipate using less than 50% of the data for this purpose.

With that in mind, the next step will be to either drop certain columns or create a new dataframe with fewer variables. Let’s get started!


```python
# Creating a new dataframe namely df.

# The dataframe will include the following columns:
cols = ['Age','Department','DistanceFromHome','Education',
         'EducationField','EmployeeCount','EnvironmentSatisfaction','Gender',
         'JobInvolvement','JobRole','JobSatisfaction','MonthlyIncome',
         'PerformanceRating','RelationshipSatisfaction','WorkLifeBalance','Attrition']

# Create DataFrame from cols above
df = data[cols]

# check the shape of the new Dataframe
print(f"The new created DataFrame(df) has {df.shape[0]} rows and {df.shape[1]} columns\n")

# Preview the new DataFrame
df.head()
```

```python
# Check missing values
df.isna().sum()
```

The dataset is complete, with no missing values.

##### Column Transformation

Several columns with integer data types require transformation, as outlined in the description above. To perform these transformations, you can use either the *map* function or the *apply* function. In this case, I’ll opt for the *apply* function.

```python

# Transform the following columns

# Education Column:  1 'Below College' 2 'College' 3 'Bachelor' 4 'Master' 5 'Doctor'
# educationMapping = {1:'Below College', 2:'College', 3:'Bachelor', 4:'Master', 5:'Doctor'}

def educationColumn(data):
    if data == 1:
        return "Below College"
    elif data == 2:
        return "College"
    elif data == 3:
        return "Bachelor"
    elif data == 4:
        return "Master"
    else:
        return "Doctor"
    

# EnvironmentSatisfaction Column: 1 'Low', 2 'Medium', 3 'High', 4 'Very High'
def environmentColumn(data):
    if data == 4:
        return "High"
    elif data == 3:
        return "High"
    elif data == 2:
        return "Medium"
    else:
        return "Low"
    
    
# JobInvolvement Column: 1 'Low', 2 'Medium', 3 'High', 4 'Very High'
def jobInvolvementColumn(data):
    if data == 1:
        return "Low"
    elif data == 2:
        return "Medium"
    elif data == 3:
        return "High"
    else:
        return "Very High"    

# JobSatisfaction Column: 1 'Low', 2 'Medium', 3 'High', 4 'Very High'
def jobSatisfactionColumn(data):
    if data == 1:
        return "Low"
    elif data == 2:
        return "Medium"
    elif data == 3:
        return "High"
    else:
        return "Very High"  
    
# PerformanceRating Column: 1 'Low', 2 'Good', 3 'Excellent', 4 'Outstanding
def performanceRatingColumn(data):
    if data == 1:
        return "Low"
    elif data == 2:
        return "Good"
    elif data == 3:
        return "Excellent"
    else:
        return "Outstanding"  

#RelationshipSatisfaction Column: 1 'Low', 2 'Medium', 3 'High', 4 'Very High'
def relationshipSatisfactionColumn(data):
    if data == 1:
        return "Low"
    elif data == 2:
        return "Medium"
    elif data == 3:
        return "High"
    else:
        return "Very High"  

# WorkLifeBalance Column: 1 'Bad', 2 'Good', 3 'Better',4 'Best'
def workLifeBalanceColumn(data):
    if data == 1:
        return "Bad"
    elif data == 2:
        return "Good"
    elif data == 3:
        return "Better"
    else:
        return "Best" 

# Call the above functions
df.loc[:,'Education'] = df.Education.apply(educationColumn)
df.loc[:,'EnvironmentSatisfaction'] = df.EnvironmentSatisfaction.apply(environmentColumn)
df.loc[:,'JobInvolvement'] = df.JobInvolvement.apply(jobInvolvementColumn)
df.loc[:,'JobSatisfaction'] = df.JobSatisfaction.apply(jobSatisfactionColumn)
df.loc[:,'PerformanceRating'] = df.PerformanceRating.apply(performanceRatingColumn)
df.loc[:,'RelationshipSatisfaction'] = df.RelationshipSatisfaction.apply(relationshipSatisfactionColumn)
df.loc[:,'WorkLifeBalance'] = df.WorkLifeBalance.apply(workLifeBalanceColumn)

df.head()
        
```

#### Number of Employees

```python
# Create a bar plot
ax = sns.countplot(data = df, x=df.EmployeeCount, palette='viridis')
plt.title('Total Number of Employees')
plt.ylabel('Frequency')
plt.xlabel('Employee Status')

# Display the count values on top of the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom',  # Center the label above the bar
                fontsize=10, color='black', 
                xytext=(0, 5),  # Offset the label slightly above the bar
                textcoords='offset points')
plt.show()
```

#### Average age of the employees

```python
avg_age = df['Age'].mean()

# Create a histogram to visualize the distribution of age
sns.histplot(df['Age'], kde=False, bins=15 )  # kde=False to remove kernel density estimation

# Add a vertical line to indicate the average age
plt.axvline(x=avg_age, color='red', linestyle='--', label=f'Mean Age: {avg_age:.2f}')
plt.title('Distribution of Employees with Average Age Line')
plt.xlabel('Age')
plt.ylabel('Count of Employees')
plt.legend()
plt.show()
```

- The visualization above indicates a normal distribution of ages, with the average age of employees in the company being 37 years.

#### No of Employees by Age Group

```python
# Define the bins (age ranges) and their corresponding labels
bins = [18, 25, 35, 45, 55, 65]  # You can adjust these ranges as needed
labels = ['18-25', '26-35', '36-45', '46-55', '56-65']

# Create a new 'AgeGroup' column based on the age bins
df.loc[:,'AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Create a count plot for the AgeGroup column
order = df["AgeGroup"].value_counts().index
ax = sns.countplot(data=df, x='AgeGroup',hue = 'AgeGroup',palette="viridis", order = order)
# Add labels on top of each bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom',  # Center the label above the bar
                fontsize=10, color='black', 
                xytext=(0, 5),  # Offset the label slightly above the bar
                textcoords='offset points')

# Add labels and title
plt.title('Distribution of Employees by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.show()
```

- While analyzing the distribution of employees by age group, it can be deduced that the majority are between 26 and 35 years old, indicating that most of the workforce consists of young adults.

#### Breakdown of distance from home by job role and attrition

```python
# Define the bins (distance ranges) and their corresponding labels
bins = [1, 6, 11, 16, 21, 26, 31]  # You can adjust these ranges as needed
labels = ['1-5', '6-10', '11-15', '16-20', '21-55', '26-30']


# Create a new 'AgeGroup' column based on the age bins
df1.loc[:,'DistanceBins'] = pd.cut(df1['DistanceFromHome'], bins=bins, labels=labels, right=False)

# Create a countplot
ax = sns.countplot(data = df1, x = 'DistanceBins', hue= 'DistanceBins')
plt.title("Attrition by Distance from Home")
plt.xlabel("Distance Range (Kms)")
plt.ylabel("Frequency")
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', 
                fontsize=10)
plt.show()
```

#### Compare average monthly income by education and attrition

```python
# Calculate average Monthly Income by Education
average_income = df1.groupby('Education')['MonthlyIncome'].mean().reset_index()

# Sort the results in descending order
sorted_education = average_income.sort_values(by='MonthlyIncome', ascending=False)

# Create a bar plot with ordered categories
ax = sns.barplot(data=sorted_education, x='Education', y='MonthlyIncome', 
                 hue='Education', errorbar=None, palette="viridis", order=sorted_education['Education'])

plt.title("Average Monthly Income by Education and Attrition")
plt.xlabel("Education Levels")
plt.ylabel("Monthly Income")

# Annotate the bars with the average income
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', 
                fontsize=10)

plt.show()
```

- The visualization on Average Monthly Income by Education and Attrition reveals that employees with doctoral degrees who attrited had the highest monthly income, followed by those with master’s degrees, while individuals with college-level education had the lowest monthly income.
